```python
# Remove duplicate rows
df = df.drop_duplicates()

# Remove rows with missing CustomerID
df = df.dropna(subset=['CustomerID'])

# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Check data after cleaning
df.info()
```

```python
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
```

```python
df.head()
```

```python
country_sales = df.groupby('Country')['TotalSales'].sum().sort_values(ascending=False)

country_sales.head(10)
```

```python
country_sales.head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Countries by Sales")
plt.xlabel("Country")
plt.ylabel("Sales")
plt.show()
```

```python
df['Month'] = df['InvoiceDate'].dt.to_period('M')

monthly_sales = df.groupby('Month')['TotalSales'].sum()

monthly_sales.plot(figsize=(12,5))

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
top_products = df.groupby('Description')['TotalSales'].sum().sort_values(ascending=False)

top_products.head(10)
```

```python
top_products = df.groupby('Description')['TotalSales'].sum().sort_values(ascending=False)

top_products.head(10).plot(kind='bar', figsize=(12,6))

plt.title("Top 10 Products")
plt.xlabel("Product")
plt.ylabel("Sales")

plt.show()
```

```python
# Return orders (Negative Quantity)
returns = df[df['Quantity'] < 0]

print("Total Return Orders:", len(returns))

returns.head()
```

```python
total_orders = len(df)
returned_orders = len(returns)

return_percentage = (returned_orders / total_orders) * 100

print(f"Return Percentage: {return_percentage:.2f}%")
```

```python
top_returns = returns.groupby('Description')['Quantity'].count().sort_values(ascending=False)

top_returns.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

top_returns.head(10).plot(kind='bar', color='red')

plt.title("Top 10 Returned Products")
plt.xlabel("Product")
plt.ylabel("Number of Returns")

plt.xticks(rotation=75)

plt.show()
```

```python
country_returns = returns.groupby('Country')['Quantity'].count().sort_values(ascending=False)

country_returns.head(10)
```

```python
plt.figure(figsize=(10,6))

country_returns.head(10).plot(kind='bar', color='orange')

plt.title("Top Countries by Returns")
plt.xlabel("Country")
plt.ylabel("Return Count")

plt.xticks(rotation=45)

plt.show()
```

```python
# Return products
returns = df[df['Quantity'] < 0]

# Top 20 returned products
high_risk_products = (
    returns.groupby('Description')
    .size()
    .sort_values(ascending=False)
    .head(20)
)

print(high_risk_products)
```
